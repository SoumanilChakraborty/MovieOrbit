```python
import pandas as pd
```

```python
creds=pd.read_csv("tmdb_5000_credits.csv")
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head(1)
```

```python
creds.head()
```

```python
movies=movies.merge(creds,on="title")
movies.head()
```

```python
#Important Features for classification:
movies = movies[["genres","id","overview","keywords","title","cast","crew"]]
```

```python
#Handling null values
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres #It is a string needs to be converted into list items.
```

```python
#FUnction to seperate the genre 
import ast
def filter(obj):
    arr=[]
    for i in ast.literal_eval(obj):
        arr.append(i['name'])
    return arr
```

```python
movies['genres']=movies['genres'].apply(filter)
```

```python
movies['keywords']=movies['keywords'].apply(filter)
```

```python
movies.head()
```

```python
#Extracting top 3 actors for each movies.
def filter2(obj):
    arr=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            arr.append(i['name'])
            c=c+1
        else:
            break
    return arr
```

```python
movies['cast']=movies['cast'].apply(filter2)
```

```python
#Extracting the directors for each movie.
def filter3(obj):
    arr=[]
    c=0
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            arr.append(i['name'])
            break
    return arr
```

```python
movies['crew']=movies['crew'].apply(filter3)
```

```python
#Converting the overview column from string to list for furthur functions.
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#For Better Tags in recommender system ,we remove the extra spaces between names,genres,cast for better tags(ex:"Christopher Nolan" will be "ChristopherNolan"):
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
upd_movies= movies[["id",'title','tags']]
```

```python
upd_movies.head()
```

```python
#Converting them back to string
upd_movies['tags']=upd_movies['tags'].apply(lambda x:" ".join(x))
```

```python
#Converting tags to lower case for better user convienience.
upd_movies['tags']=upd_movies['tags'].apply(lambda x:x.lower())
```

```python
upd_movies.head(9)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
#Vectorization of tags column , also being converted to array for simplicity as it gives a default scipy sparse matrix output
vector_tags=cv.fit_transform(upd_movies['tags']).toarray()
vector_tags
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stemming(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    #converting string to list
    return " ".join(y)
```

```python
upd_movies['tags']=upd_movies['tags'].apply(stemming)
```

```python
#Calculation of cosine distance between vector tags
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector_tags)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x : x[1])[1:6]
```

```python
def similar_recommend(movie):
    movie_index=upd_movies[upd_movies['title'] == movie].index[0]
    distances=similarity[movie_index]
    distances
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:11]
    for i in movie_list:
        print(upd_movies.iloc[i[0]].title)
```

```python
similar_recommend('Thor')
```
